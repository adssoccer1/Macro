### Emerging Markets
People invest in EMs like they are all one country, even though individual country's conditions are radically different. EM through time is a good alpha landscape. They treat EM as one homogenous asset class. 

BRICS used to be what FAANG is today. When expectations for BRICS was disapointed capital left. 

With regards to changes in monetary flows, when QE stops and there is QT or rate hikes, the assets that were receiving the easy flows get hit the hardest, the flows do not necessarily leave EM. 

A lot of trading is narrative driven and backwards looking. Instead one needs to ask "why does that thing matter? Which flow does it connect to? Is it a large flow? Where is that flow going? Which asset is it supporting? And when does that flow stop?"

Totem has a framework which is a money and credit flows framework which lets them test out what is driving a specific flow. They have a process for taking in news flow accross the 35 countries x 6 asset classes every day. There is a lot of narrative that are actually very small flows relative to actuall flow with real size like pension fund movements that drive the trend. 

""People tend to focus on these week to week narratives, but broadly speaking these things do not matter. So what totem tries to do is to think at the highet level first. Money creation works through financial channels into assets, credit works through economic channels. So okay which players are taking their investments and putting them into which assets and what are driving those players?"" 

If we look at something like NVDIA for example, we can say okay retail is the one driving this rally so its a retail flow then what will stop this buying? well the halting of the expansion of the Fed balance sheet will stop this flow. 

Another example would be emerging market stocks. A lot of the flow that goes into EM stocks is from foreigners that look for momentum. So cross-border flow chases what has done well, so what you see is an asset that is running hot where there is a current account deficit, fiscal deficit and the currency is expensive and depending on foreign flows. This was basically EMs in 2011, or BRICS in 2011. But then when this foreign flow of money stops to these countries, you begin to see a balance of payments crisis. The currency falls, the rates goes up, fiscal tightens, private credit creation falls, the economy weakens, import spending goes down, and the current account closes. 

So if a country has a large current account which requires a lot of foreign investment, but that gap closes, then once this process troughs the currency will stabalize, inflation will come down, rates can ease, credit eases, economies rebound. So thats what the beginning of the rally in these EMs looks like. Then foreigners see this momentum begin to pick up and they chase these returns.
