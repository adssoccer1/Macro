```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1) Define the months in ascending order (Feb 2023 → Nov 2024)
months = [
    "Feb 2023", "Mar 2023", "Apr 2023", "May 2023", "Jun 2023", "Jul 2023",
    "Aug 2023", "Sep 2023", "Oct 2023", "Nov 2023", "Dec 2023", "Jan 2024",
    "Feb 2024", "Mar 2024", "Apr 2024", "May 2024", "Jun 2024", "Jul 2024",
    "Aug 2024", "Sep 2024", "Oct 2024", "Nov 2024"
]

# 2) Paste your data from bottom row (Feb 2023) to top row (Nov 2024)
#    Format: (total_holdings, net_sales, valuation_change)
#    A positive net_sales means net buying (bar in green).
data = [
    (770247,  -13261, -22629),  # Feb 2023
    (793206,    4800,   17510), # Mar 2023
    (798203,   -2019,    7180), # Apr 2023
    (794713,   -5180,     476), # May 2023
    (837510,   -1009,   41103), # Jun 2023
    (854424,   -9293,   27871), # Jul 2023
    (838345,   -2734,  -15317), # Aug 2023
    (806594,    2648,  -33169), # Sep 2023
    (790570,    1422,  -16387), # Oct 2023
    (850432,   -3488,   62414), # Nov 2023
    (887194,   -6623,   42449), # Dec 2023
    (905713,    -845,   23168), # Jan 2024
    (939382,   -3143,   34562), # Feb 2024
    (954252,  -10089,   23765), # Mar 2024
    (916347,   -2346,  -36966), # Apr 2024
    (955353,   -1738,   41273), # May 2024
    (976819,   -4277,   22030), # Jun 2024
    (989595,    -997,   12721), # Jul 2024
    (1019662,   8665,   20836), # Aug 2024
    (1041825,   5884,   17234), # Sep 2024
    (1038157,  -1435,   -2311), # Oct 2024
    (1093989,  -3082,   58629), # Nov 2024
]

# 3) Convert the data into a DataFrame
df = pd.DataFrame(data, columns=["TotalHoldings", "NetSales", "ValChange"], index=months)

# 4) Create the plot
plt.figure(figsize=(10, 6))
x = np.arange(len(months))

# Plot the line for total holdings (Japan's holdings of U.S. equities)
plt.plot(
    x, 
    df["TotalHoldings"], 
    color="black", 
    linewidth=2,
    marker="o",
    label="Total Holdings (Japan)"
)

# Determine the color for Net Sales: green if ≥ 0, red if < 0
net_sales_colors = ["green" if val >= 0 else "red" for val in df["NetSales"]]
# Determine the color for Valuation Change: blue if ≥ 0, orange if < 0
val_change_colors = ["blue" if val >= 0 else "orange" for val in df["ValChange"]]

bar_width = 0.35

# Plot the Net Sales bars
plt.bar(
    x - bar_width/2, 
    df["NetSales"], 
    width=bar_width, 
    color=net_sales_colors, 
    alpha=0.7, 
    label="Net Sales (Green=Net Buying, Red=Net Selling)"
)

# Plot the Valuation Change bars
plt.bar(
    x + bar_width/2, 
    df["ValChange"], 
    width=bar_width, 
    color=val_change_colors, 
    alpha=0.7, 
    label="Valuation Change (Blue=Positive, Orange=Negative)"
)

# 5) Formatting
plt.title("Japan's Holdings of U.S. Corporate Equities: Net Sales & Valuation Changes")
plt.xticks(x, months, rotation=45)
plt.xlabel("Month")
plt.ylabel("Amount Billions of USD")
plt.legend()
plt.tight_layout()
plt.show()
```
