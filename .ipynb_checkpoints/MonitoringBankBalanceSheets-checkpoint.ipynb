# Monitoring Bank Balance Sheets

Banks can create money via private loans and extend leverage via secured financing.

When Commercial banks make loans they digitally print money by inserting **newly created** bank deposits into private customer accounts. An asset and a liability is created on the bank balance sheet and the total money supply increases by the amount of the newly created liability. 

When banks extend secured leverage, they move **existing** bank reserves around to the accounts with the highest utility for them. Repo markets are the lifeblood of the banking ecosystem so what happens if SLRs begin affecting repo activity? Equity lending and borrowing can be key markets used by Hedge Funds and other equity market participants to go levered long or short equities. What happens when equity lending/borrowing slows or accelerates?

Bank balance sheets dictate the amount of leverage and credit they can extend, thus we must monitor when balance sheets infringe on a bank's ability to function. The Supplementary Leverage Ratio (SLR) and the Liquidity Coverage Ratio (LCR) are crucial and provide a ceiling and a floor, respectively, to the amount of risk a bank can extend. 

## Private Lending
Pre-GFC, the money supply (M2) was highly correlated with private lending. Post GFC, in the era of QE and large deficits, government, not commercial banks, has been a more influential in money supply growth. While QE does increase bank reserves which are not part of m2, when the Fed buys bonds from a non-bank entity, they inject bank deposits into M2. Fiscal deficits mean the Treasury issues treasuries and swap central bank reserves for treasuries on commercial bank balance sheets. Then when the TGA is spent down, they inject those deposits back into the private sector esentially converting commerical bank reserves (not part of M2) into bank deposits. While the newly added treasuries are highly money like, they are not included in M2.


Many anticipate [the monetary stimulus and financial engineering](https://github.com/adssoccer1/Macro/blob/main/LiquidityInjections.ipynb) that has occured since the GFC to subside in the near future, bringing financial assets and the money supply into closer contact with the real economy and private lending.

## Secured Financing
Unlike private lending, secured financing transactions do not create new net financial assets but instead involve the temporary reallocation of assets/liabilities that already exist, and therefore, do not expand/contract the balance sheet. 

Given the secured financing ecosystem involves many players like commercial banks, primary dealers, prime brokers, money market funds, and shadow banks, it is difficult to capture granular detail on secured financing in a timely manner. One must balance a collection of indicators to derive insights. 

The Federal Reserve Bank of New York offers weekly Primary Dealer statistics which is a timely proxy for the rest of the system. Primary Dealer balance sheet snapshots give us insight into repo markets and equity lending. Both of these markets are frequently tapped for leverage by investors. If a hedge fund has an equity position, but wants to increase risk, they can lend equities they own via a primary dealer for cash and deploy that cash into more equities. 

Coupling indicators like primary dealer equity lending/borrowing activity and funding spreads like equity financing costs to relative to SOFR or IORB can provide compelling insight into the state of leverage being used by market participants to buy/sell stocks.

Observe the below Primary Dealer equity lending vs borrowing spread. If participants are adding long equity exposure, they can lend equities to primary dealers for cash and buy more equities. 	If participants are short equities, they will borrow shares—potentially from primary dealers or other lenders—sell them in the market, and later repurchase them at a lower price to return to the lender at the end of the equity lending agreement. In the charts below you can see large spreads during aggressive bull markets like 2023 and 2024. 


Primary dealers represent a portion of total repo activity and have limited involvement in equity lending so they are only a proxy. Prime brokers are more commonly used for equity lending. I will break down the specific sizes of relative repo and equity lending markets and their participants in another piece.

More general banking liquidity conditions can be observed by looking at primary dealer balance sheet composition. For example, if one observes a high concentration of treasuries held by primary dealers, but a decrease in primary dealer repo activity could signal primary dealers are absorbing treasury supply, but unable to finance it via repo, implying balance sheet constraints. The SLR requires banks to maintain 3% and Global Systemically Important Banks (GSIBs) to maintain 5% of the following equation: SLR = Tier 1 Capital / Total Leverage Exposure. 

Central bank reserves and Treasuries are assets that count towards Total Leverage Exposure, so if a bank ingests too many Treasuries, it will limit how much repo that can extend. I mentioned earlier that secured financing activities like repo transactions do not actually expand bank balance sheets or increase M2, and this is partly true. They do not create new assets and liabilities on balance sheets like private lending, but they do still count as leverage exposure under the SLR ratio. 

Therefore bank balance sheets can actually absorb too many "safe" assets like central bank reserves and treasuries that restricts their ability to provide leverage. The SLR ratio creates a ceiling on the amount of money in the financial system that is ideal for continued banking activity. 

If the SLR creates a ceiling on the optimal amount of HQLA in the banking system, the LCR requirements create a floor. The LCR requires banks to hold sufficient HQLA to meet 30 day cash outflows. The Fed constantly evaluates the “lowest comfortable level of reserves” (LCLoR) in the system. If there are not enough reserves, liquidity constraints can emerge, increasing short-term funding stress, but if there are too many reserves in the banking system, the SLR begins to drag leverage expansion. So why not simply remove central bank reserves and treasury assets from the SLR calculations to allow for more balance sheet expansion... perhaps Bessent and Trump will do just that. 

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# API URLs
urls = [
    # Securities In (Borrowed)
    "https://markets.newyorkfed.org/api/pd/get/SBN2013/timeseries/PDSIOSB-E_PDSIOSB-ETAL30_PDSIOSB-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2015/timeseries/PDSIOSB-E_PDSIOSB-ETAL30_PDSIOSB-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2022/timeseries/PDSIOSB-E_PDSIOSB-ETAL30_PDSIOSB-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2024/timeseries/PDSIOSB-E_PDSIOSB-ETAL30_PDSIOSB-ETAG30.json",
    # Securities Out (Lent)
    "https://markets.newyorkfed.org/api/pd/get/SBN2024/timeseries/PDSOOS-E_PDSOOS-ETAL30_PDSOOS-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2022/timeseries/PDSOOS-E_PDSOOS-ETAL30_PDSOOS-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2015/timeseries/PDSOOS-E_PDSOOS-ETAL30_PDSOOS-ETAG30.json",
    "https://markets.newyorkfed.org/api/pd/get/SBN2013/timeseries/PDSOOS-E_PDSOOS-ETAL30_PDSOOS-ETAG30.json"
]

def fetch_data_from_url(url):
    """Fetch JSON data from the given URL and return a DataFrame."""
    response = requests.get(url)
    if response.status_code == 200:
        json_data = response.json()
        # Extract the timeseries data
        data = [
            {
                "date": entry["asofdate"],
                "keyid": entry["keyid"],
                "value": entry["value"] if entry["value"] != "*" else None
            }
            for entry in json_data["pd"]["timeseries"]
        ]
        return pd.DataFrame(data)
    else:
        print(f"Failed to fetch data from {url}, status code: {response.status_code}")
        return pd.DataFrame()

# Aggregate data from all URLs
all_data = pd.DataFrame()
for url in urls:
    df = fetch_data_from_url(url)
    all_data = pd.concat([all_data, df], ignore_index=True)

# Data cleaning
all_data["date"] = pd.to_datetime(all_data["date"])
all_data["value"] = pd.to_numeric(all_data["value"], errors="coerce")  # Convert "*" to NaN
all_data.dropna(subset=["value"], inplace=True)

# Pivot the data for visualization
pivot_data = all_data.pivot(index="date", columns="keyid", values="value")

# Rename columns for better understanding
pivot_data.rename(
    columns={
        "PDSIOSB-E": "Total Equities In",
        "PDSIOSB-ETAG30": "Equities In Under Term Agreements (>30 days)",
        "PDSIOSB-ETAL30": "Equities In Under Term Agreements (<30 days)",
        "PDSOOS-E": "Total Equities Out",
        "PDSOOS-ETAG30": "Equities Out Under Term Agreements (>30 days)",
        "PDSOOS-ETAL30": "Equities Out Under Term Agreements (<30 days)"
    },
    inplace=True,
)

# First Chart: All categories (as before)
plt.figure(figsize=(12, 6))
for column in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[column], label=column, marker="o")

# Customize the x-axis to show one year at a time
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1))  # Set major ticks to every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format ticks as years

# Customize the chart
plt.title("Equities Lending and Borrowing Activity by Primary Dealers", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Value (in Millions)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the first plot
plt.show()

# Second Chart: Total Securities In vs. Total Securities Out
plt.figure(figsize=(12, 6))
if "Total Equities In" in pivot_data.columns and "Total Equities Out" in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data["Total Equities In"], label="Total Equites In", marker="o")
    plt.plot(pivot_data.index, pivot_data["Total Equities Out"], label="Total Equities Out", marker="x")

# Customize the x-axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Customize the chart
plt.title("Primary Dealer Total Equities In vs. Out Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Value (in Millions)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the second plot
plt.show()
```
