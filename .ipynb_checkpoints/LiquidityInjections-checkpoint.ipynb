# Liquidity Injections

The post-Great Financial Crisis era has seen incredible performance from financial assets in the United States, primarily because, at any given time, there has been a steady injection of liquidity from U.S. monetary and/or fiscal authorities.

Spending in an economy can be driven by credit growth or money printing. Credit growth occurs when commercial banks lend to a private sector that is productive enough to use credit to add to the exchange of goods and services. This directly drives real economic growth.  Money printing occurs when central banks create bank reserves that float asset prices before stimulating an increase in the exchange of goods and services. Money printing stimulates spending by pushing risk further out the risk curve and through wealth effects that precede changes in real economic activity.


Since the GFC, credit growth has stalled, yet the consumer, supported by strong income growth, has maintained a strong level of spending (both at a healthy ~5%).



Pre-GFC, households levered their balance sheets with credit to support their consumption, but today, we seem to be in an income-driven expansion supported by ever-increasing USD asset prices, not credit creation.



In a hyperfinancialized economy, U.S. corporations can hire more workers because their stocks keep rising and it’s cheap to issue debt! Households can maintain low saving rates (high spending rates) because the value of their house and 401(k) keep going up!

It is thus crucial to maintain the USD asset bubble; otherwise, the income-driven expansion breaks. The Monetary and Fiscal powers know this.

In 2025, a looming debt maturity wall, a likely spend-down and subsequent replenishing of the TGA, and a potentially hawkish Treasury Secretary Bessent who may move issuance to the long end of the curve, threaten the USD asset bubble and necessitate liquidity injections to prop up already elevated financial assets.

Let’s look at past liquidity-injecting tools, both explicit and discreet, that monetary and fiscal authorities have used to float asset prices since the GFC, what they may use in 2025, and why, if no liquidity is injected from this point on, the second half of 2025 may see a correction.

Fiscal deficits are stimulative because:
1. The Treasury issues Treasuries and receives central bank reserves to fill the Treasury General Account (TGA).
2. Commercial banks purchase Treasuries by exchanging their central bank reserves for interest-bearing Treasuries.
3. The Treasury disburses the new reserves in the TGA to:
    - Households: Through programs like stimulus checks, Social Security payments, and unemployment benefits.
    - Businesses: Via government programs like the Paycheck Protection Program (PPP) or contracts.
    - State and Local Governments: Through grants and federal funding.
    - These recipients deposit funds into commercial bank accounts, which increases the central bank reserves of those banks at the central bank.

Quantitative Easing is stimulative because:
1. Purchasing Treasuries and MBS reduces their supply, which reduces rates across the curve.
2. Lower rates force sellers of Treasuries to reinvest capital further out the risk curve into higher-yielding assets like corporate bonds or equities.

## More Specifically

The formality and size of QE and fiscal programs make them easy to track and anticipate. Along with “ZIRP,” these tools have laid the foundation for the income-driven expansion we find ourselves in today.

However, since the GFC, a wide range of tools have been used to stabilize liquidity, dampen economic volatility, and prolong the USD asset bubble. When considering a trade, one must monitor the following tools and weigh their potential impact against one's thesis and what is currently priced.

- In 2008, **dollar-swap lines** were established between the Fed and other major central banks to provide dollar liquidity that those central banks could disperse to foreign commercial banks during times of dollar liquidity stress. Primarily used in crisis scenarios, swap lines are an example of monetary powers dampening volatility. With dollar swap lines, foreigners know they can access dollars and avoid fire-selling their USD assets during a crisis.

*The Fed charges interest on the dollar loans, typically based on the Overnight Index Swap (OIS) rate plus a small margin.
[link](https://fred.stlouisfed.org/series/WCBLSA)


- In 2011, "[Operation Twist](https://www.federalreserve.gov/newsevents/pressreleases/monetary20110921a.htm)" was introduced to flatten the yield curve without expanding the Fed’s balance sheet. The Fed sold Treasuries on the front end of the curve and bought Treasuries on the long end, effectively lowering long-term rates. This type of operation comes with a formal announcement, making it relatively easy to anticipate.

- In 2019, the private repo markets experienced a spike in overnight rates, rising from 2% to as high as 5%. This created a massive spread against the Fed Funds rate, signaling stress in funding markets. To address this, the Fed initiated open market operations in the repo markets, injecting liquidity to prevent fire sales by cash-strapped banks needing to meet settlement or funding obligations. In 2021, the Fed formally announced the **Standing Repo Facility (SRF)** so primary dealers and depository institutions could deplosit assets as collateral for funding from the Fed. There is a stigma associated with using the SRF as it indicates funding stress at the respective institution, but the SRF is fundementally another volatility dampener to liquidity shortages in the intra banking system. Monitoring spreads like SOFR vs EFFR is crucial. While limited volume in the SRF does not necessarily indicate ample liquidity, inceases in volume at the SRF do indicate liquidity stress.  

 
[link](https://fred.stlouisfed.org/series/RPONTSYD)



- In 2023 and 2024, the **Reverse Repo Facility (RRP)** was a key source of liquidity. While the RRP is traditionally a tool to drain excess liquidity in repo markets and maintain a floor on the Fed Funds rate, it can also act as a source of liquidity when funds in the RRP flow back into private repo markets.

[link](https://fred.stlouisfed.org/graph/?g=EPMW)

- In 2023, Treasury Secretary Janet Yellen decided to finance the Treasury by issuing more T-bills to avoid crowding out other investments. She recognized that money from Money Market Funds sitting in the RRP would be redirected to purchase the newly issued T-bills, effectively moving $500 billion out of the RRP and into the base money supply.

- In response to the SVB crisis, the **Bank Term Funding Program (BTFP)** allowed vulnerable banks with unrealized bond losses, caused by the fastest Fed hiking cycle in history, to borrow from the Fed using Treauries as collateral at par value. The speed of the BTFP roll out highlights the "Fed Put" in action. In a sense, the BTFP was the SRF on steriods as it allowed for lending terms up to 1 year (compared to the SRF’s overnight terms) and collateral valuation was at par despite heavy unrealized losses. This was the Fed Put in action, preventing fire sales that would have taken down the financial system in a different era.

- Though the BTFP ceased lending in 2024, the **Fed Discount Window** remains a permanent source of liquidity and a strong signal for banks in distress. Unlike the SRF, Discount Window lending is longer term, but like the SRF, limited volume does not necessarily indicate ample liquidity, while increases in volume signal liquidity stress. **Psychologically, the BTFP and the Discount Window reinforce market confidence in the power of the Fed Put**. 

  [link](https://fred.stlouisfed.org/series/BOGZ1FA713068703Q)


- The **Treasury General Account (TGA)** is another must watch piece of the liquity puzzle. From 2008–2015, the Treasury began building up the TGA by issuing Treasuries, a liquidity drain that was offset by ongoing QE. Current policy requires the TGA to maintain one week’s worth of outflows as a hedge against scenarios where the Treasury cannot issue new debt to meet obligations. When the Treasury disburses funds from the TGA (e.g., for government spending), these payments increase bank reserves and inject liquidity into the financial system. Thus, a TGA spend-down, without immediate replenishment, is liquidity additive. In 2025, another debt ceiling standoff is prohibiting the Treasury from issuing new debt, draining the TGA without replenishment.

- Does the Trump administration secretly want a debt ceiling standoff to trigger a TGA spend-down that supports markets as he begins his term? Moreover, how will Treasury Secretary Bessent fund the replenishment of the TGA in the second half of this year, considering his criticism of Yellen’s reliance on bill issuance?
