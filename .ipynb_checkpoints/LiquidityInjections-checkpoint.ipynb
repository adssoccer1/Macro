# Liquidity Injections

The post-Great Financial Crisis era has seen incredible performance from financial assets in the United States, primarily because at any given time there has been a steady injection of liquidity from U.S. monetary and/or fiscal powers. 

Spending in an economy can be driven by credit growth or money printing. Credit growth occurs when commercial banks lend to a private sector that is productive enough to use credit in the exchange of goods and services. This directly drives real economic growth. Money printing occurs when central banks create bank reserves that float financial asset prices **before** stimulating an increase in the exchange of goods and services, AKA the real economy. Money printing stimulates spending by pushing risk further out the risk curve and through wealth effects, that preceede changes in spending habbits.

Since the GFC credit growth has stalled, yet the consumer, supported by strong income growth, has maintained a strong level of spending (both at a healhty ~5%). 



Pre GFC, households levered their balance sheets with credit to support their consumption, but today we seem to be in an income driven expansion that is supported by ever-increasing USD asset prices, not credit creation. 



In a hyperfinancialized economy, U.S. corporations can hire more workers because their stocks keep rising and it's cheap to issue debt! Households can maintain low saving rates (spend)  because the value of their house and 401k keep going up!

It is thus crucial to maintain the USD asset bubble, else the income driven expansion breaks. The Fed knows this. 

In 2025, a looming debt maturity wall, a likely spend-down and proceeding replenishing of the TGA, and a potentially hawkish Treasury Secretary Bessent that may move issuance to the long end of the curve, threaten the USD asset bubble and beg [USE A BETTER WORD OR PHRASE] liquidity injections into the financial system to prop up elevated financial assets.

 Let's look at past liquidity injecting tools, both explicit and discreet, that the monetary and fiscal powers have used to float asset prices since the Great Financial Crisis, what they may use in 2025, and why if no liquidity is injected from this point on, the second half of 2025 will likely see a correction.

Fiscal deficits are stimulative because:
1. The Treasury issues Treasuries and receives central bank reserves to fill the Treasury General Account (TGA).
2. Commercial banks purchase Treasuries by exchanging their central bank reserves for interest-bearing Treasuries.
3. The Treasury disburses the new reserves in the TGA to:
    - Households: Through programs like stimulus checks, Social Security payments, and unemployment benefits.
    - Businesses: Via government programs like the Paycheck Protection Program (PPP) or contracts.
    - State and Local Governments: Through grants and federal funding.
    - These recipients deposit funds into commercial bank accounts, which increases the central bank reserves of those banks at the central bank.

Quantitative Easing is stimulative because:
1. Purchasing Treasuries and MBS reduces their supply, which reduces rates across the curve.
2. Lower rates force sellers of Treasuries to reinvest capital further out the risk curve into higher-yielding assets like corporate bonds or equities.

## More Specificaly

The formality and size of QE programs and Fiscal bills make them easy to track and anticipate, but since the GFC a wide range of tools have been used to stabalize liquidity to deflect any needles that may come close to popping the USD asset bubble. 

- In 2008 Dollar-swap Lines were established between the Fed and other major Central Banks to provide dollar liquidty that those Central Banks could disperse these dollar to their commercial banks in times of dollar liquidity stress. While this measure is primarily used in crisis scenarios, its an example of monetary powers dampening volatility. With dollar swap lines, Foreigners know they can get access to dollar and do not need to fire sell their USD assets in a crisis. 

*The Fed charges interest on the dollar loans, typically based on the Overnight Index Swap (OIS) rate plus a small margin.
[link](https://fred.stlouisfed.org/series/WCBLSA)




- In 2011 "[Operation Twist](https://www.federalreserve.gov/newsevents/pressreleases/monetary20110921a.htm)" was purposed to flatten the yeild curve, but without balance sheet expansion. The Fed sold treasuries on the front end of the curve and bought treasuries on the the long end, lowering long term rates. This kind of operation comes with a formal announcement, of course and is therefore easy to anticipate.

- In 2019, the private repo markets saw a spike in overnight rates from 2% to as high as 5%, creating a massive spread between the Fed Funds rate. The Fed began open market operations in the repo markets providing liquidity for assets, preventing any fire sales from cash strapped banks that needed to meet settlement/funding obligations.  In 2021 the Fed formally announced the Standing Repo Facility (SRF) so primary dealers and depository institutions could deplosit assets as collateral for funding from the Fed. There is a stigma associated with using the SRF as it indicates funding stress at the respective institution, but the SRF is fundementally another volatility dampener to liquidity shortages in the intra banking system. Monitoring spreads like SOFR vs EFFR is crucial. While limited volume in the SRF does not necessarily indicate ample liquidity, inceases in volume at the SRF do indicate liquidity stress.  

 
[link](https://fred.stlouisfed.org/series/RPONTSYD)



- In 2023 and 2024 the Reverse Repo Facility (RRP) was actually a key source of liquidity. While the RRP is a tool to drain excess liquidty in repo markets and maintain a floor on the Fed Funds rate, it can be a source of liquidity when funds in the RRP move back into private repo markets.

[link](https://fred.stlouisfed.org/graph/?g=EPMW)


- In 2023 in response to the SVB crisis, the ank Term Funding Program (BTFP) allowed vulnerable banks with unrealized bond losses as a result of the fastest Fed hiking cycle in history, to borrow from the Fed using those treauries as collateral at par value. The speed at which the BTFP rolled out highlights the Fed Put in action. In a sense the BTFP was the SRF on steriods as it allowed for lending terms up to 1 year vs overnight and collateral valuation was at par despite heavy unrealized losses. This was the Fed Put in action , preventing fire sales that would have taken down the financial system in a different era.

- Though the BTFP ceased lending in 2024, the Fed Discount Window is a permenant source of liquidity and a strong signal for banks in distress. Unlike to the SRF the lending is longer term, but similarily limited volume in the Discount Window does not necessarily indicate ample liquidity, inceases in volume at the SRF do indicate liquidity stress. **Pyschologically, the BTFP and the Discount Window gives markets confidence in the Fed Puts power**.

- The Treasury General Account (TGA) is another must watch piece of the liquity puzzle. From 2008-2015 the treasury began building up the TGA by issuing treasuries, a liquidity drain that was offset by ongoing QE. The current policy is for the TGA to maintian 1 weeks worth of outflows to hedge against a scenario where they cannot issue treasuries to meet obligations.

When the Treasury disburses funds from the TGA (e.g., for government spending), these payments increase bank reserves and inject liquidity into the financial system. So a TGA spendown, without an immediate replenishing, is liquidity additive. In 2025 we are seeing another Debt Ceiling standoff which prohibuts the treasury from issuing new debt, thereby draining the TGA without replenishment.
